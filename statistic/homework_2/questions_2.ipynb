1. Предположим, у нас есть выборка, где выборочное среднее равняется 10. Одно из наблюдений равняется 100. Что мы можем сказать об этом наблюдении?
    - Данное наблюдение точно является выбросом, т.к. очень сильно отклоняется от среднего значения
    - Такое наблюдение не может принадлежать выборке со средним значением равным 10, так как в 10 раз больше, чем выборочное среднее, вероятность такого события стремится к нулю
    - **Прежде чем судить о том, насколько необычным является это наблюдение, необходимо знать, чему равняется стандартное отклонение**

2. Если значение стандартного отклонения позволяет нам сказать, что значение 100 все таки является выбросом, что следует сделать с такими наблюдением?
    - Нельзя так просто удалять значения из выборки, оставим как есть
    - Выбросы это плохо, нужно удалить
    - **Нужно попробовать понять природу этого наблюдения (напр. аномалии при сборе данных), и уже после этого принять решение**
