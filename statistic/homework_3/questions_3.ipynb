4. Рассчитайте 99% доверительный интервал для следующего примера:

x = 10
sd = 5
n = 100

- **[8,71 ; 11,29]**

- [9,71 ; 12,29]

- [6,71 ; 13,29]

- [7,71 ; 12,29]

```python
import numpy as np
import scipy.stats as st
import statsmodels.stats.api as sms

mu = 10
sd = 5
n = 100
se = sd/np.sqrt(n)
se
```

```python
bound = se * 2.58
```

```python
upper = mu + bound
lower = mu - bound
print(lower, upper)
```

6. При p < 0.05 нулевая гипотеза обычно отклоняется, и различия считаются статистически достоверными.  
Однако часто в статистике используется более жесткий критерий достоверности различий, например, при условии, что p < 0.01p<0.01.

Значение p-уровня значимости, которое выбирается в качестве порога, обозначается буквой \alphaα (альфа). Например, если исследователь решил, что \alpha = 0.05α=0.05, то и нулевая гипотеза будет отклоняться при условии, что p < 0.05p<0.05. 

Если в определенной ситуации весьма рискованно отклонить нулевую гипотезу,
когда она на самом деле верна, то лучше использовать показатель \alphaα равный:


- 0.5

- 0.05

- 0.1

- **0.001**

Чем меньше альфа тем лучше
